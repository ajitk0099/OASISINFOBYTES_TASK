```python
# Boxplot  Diagram Using matplotlib & Seaborn libraries
```

```python
plt.figure(figsize=(12,8))
columns={0:"TV",1:"Radio",2:"Newspaper",3:"Sales"}
for plot,col_name in columns.items():
    plt.subplot(2,2,plot+1)
    sns.boxplot(data=data[col_name],palette="pastel")
    plt.xlabel(col_name)
    plt.title("Outliers in {} feature".format(col_name))
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,12))
columns={0:"TV",1:"Radio",2:"Newspaper",3:"Sales"}
for plot,col_name in columns.items():
    plt.subplot(4,4,plot+1)
    sns.boxplot(data=data[col_name],palette="pastel")
    plt.xlabel(col_name)
    plt.title("Outliers in {} feature".format(col_name))
    
plt.tight_layout()
plt.show()
```

```python
# Distribution plots 
```

```python
plt.figure(figsize=(15, 5))

columns = {0: "TV", 1: "Radio", 2: "Newspaper"}

for plot, col_name in columns.items():
    plt.subplot(1, 3, plot + 1)
    sns.distplot(x=data[col_name],hist=False, label= col_name)
    sns.distplot(x=data["Sales"],hist = False,label= "Sales")
    plt.xlabel("{} VS Sales".format(col_name))
    plt.title("Distribution plot {} VS Sales".format(col_name))
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

columns = {0: "TV", 1: "Radio", 2: "Newspaper"}

for plot, col_name in columns.items():
    plt.subplot(1, 3, plot + 1)
    sns.scatterplot(x=data[col_name],y=data["Sales"], label= col_name,palette="pastel")
    plt.xlabel("{} VS Sales".format(col_name))
    plt.title("Scatter plot: {} VS Sales".format(col_name))
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data,x="TV",y="Sales",style="Radio",size="Newspaper",palette="pastel",legend=False)
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("Relationship between TV and Sales with Radio Impact and Newspaper")
plt.show()
```
